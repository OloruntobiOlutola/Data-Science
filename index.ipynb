```python
root_url = 'https://www.football-data.co.uk/mmz4281/'
frames = []
league_dictionary = {
    'Spanish La Liga':'SP1',
    'German Bundesliga':'D1', 
    'English Premier League':'E0',
}
seasons = ['0506', '0607', '0708', '0809', '0910', '1011', '1112', '1213', '1314', '1415', '1516', '1617', '1718', '1819', '1920', '2021', '2122', '2223', '2324', '2425']
```

```python
count = 0
for league in league_dictionary:
    for season in seasons:
        count += 1
        print(league_dictionary[league], '-', season,  count)
```

```python
dict_historical_data = {}
for league in league_dictionary:
    frames = []
    for season in seasons:
        url = root_url + season + '/' + league_dictionary[league] + '.csv'
        df = pd.read_csv(url, encoding='unicode_escape')
        df.insert(0, 'season', season[:2] + '/' + season[2:])
        frames.append(df)
        dict_concat = pd.concat(frames)
    dict_historical_data[league] = dict_concat
```
