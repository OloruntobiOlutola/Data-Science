# Web Scrapping

```python
df_premier24 =  pd.read_csv('https://www.football-data.co.uk/mmz4281/2425/E0.csv')
```

```python
df_premier24
```

```python
df_premier24.rename(columns={'Date':'date',
                                            'HomeTeam':'home_team',
                                            'AwayTeam':'away_team',
                                            'FTHG': 'home_goals',
                                            'FTAG': 'away_goals'}, inplace=True)
```

```python
df_premier24
```

```python
root_url = 'https://www.football-data.co.uk/mmz4281/'
frames = []
league_dictionary = {
    'Spanish La Liga':'SP1', 
    'Spanish Segunda Division':'SP2',
    'German Bundesliga':'D1', 
    # 'English Premier League':'E0',
}
seasons = ['0506', '0607', '0708', '0809', '0910', '1011', '1112', '1213', '1314', '1415', '1516', '1617', '1718', '1819', '1920', '2021', '2122', '2223', '2324', '2425']
```

```python
for league in league_dictionary:
    print(league_dictionary[league])
```

```python
for season in seasons:
    for league in leagues:
        url = root_url + season + '/' + str(league_dictionary[league]) + '.csv'
        df = pd.read_csv(url, encoding='unicode_escape')
        df.insert(1, 'season', season[:2] + '/' + season[2:])
        frames.append(df)
```

```python
len(frames)
```
