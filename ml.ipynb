```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot  as plt
from sklearn import linear_model
```

```python
df_boston = pd.read_csv('Boston House Prices.csv')
```

```python
df_boston.describe()
```

```python
import statsmodels.api as sm
```

```python
y = df_boston['Value'] # dependent variable
x = df_boston['Rooms'] # independent variable
```

```python
df_boston.plot(kind='scatter', x= 'Rooms', y='Value')
```

```python
df_boston.plot(kind='scatter', x= 'Distance', y='Value')
```

```python
x = sm.add_constant(x)
```

```python
lm = sm.OLS(y,x).fit()
```

```python
lm
```

```python
lm.predict(x)
```

```python
lm.summary()
```

```python
# Rooms coef: 9.1021
# Constant coef: - 34.6706

# Linear equation: 𝑦 = 𝑎𝑥 + 𝑏
y_pred = 9.1021 * x['Rooms'] - 34.6706
```

```python
plt.figure(figsize=(6,4), tight_layout=True)
sns.scatterplot(x=x['Rooms'], y=y)

sns.lineplot(x=x['Rooms'], y=y_pred, color='red')

plt.xlim(0)
plt.ylim(0)
plt.savefig('linear_regression')
plt.show()
```

```python
y = df_boston['Value'] # dependent variable
X = df_boston[['Rooms', 'Distance']] # independent variable
```

```python
X = sm.add_constant(X) # adding a constant
lm = sm.OLS(y, X).fit() # fitting the model
lm.summary()
```

```python
lm = linear_model.LinearRegression() 
lm.fit(X, y)
```

```python
lm.predict(X)[:5]
```

```python
df_boston
```

```python
lm.score(X, y)
```
