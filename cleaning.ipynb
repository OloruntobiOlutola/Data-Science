```python
df_netflix = pd.read_csv('netflix_titles.csv')
```

```python
df_netflix.dropna(subset=['director'])
```

```python
mode = ''.join(df_netflix['rating'].mode())
df_netflix_copy = df_netflix.copy()
```

```python
df_netflix_copy['rating'].fillna(mode, inplace=True)
```

```python
df_netflix_copy.isnull().mean()
```

```python
df_netflix_copy['duration'].fillna('0', inplace=True)
```

```python
df_netflix_copy.fillna(method="bfill")
```

```python
df_netflix_copy['type'].value_counts()
```

```python
df_netflix_copy['type'] == 'Movie'
```

```python
df_movie = df_netflix_copy[df_netflix_copy['type'] == 'Movie']
```

```python
df_movie
```

```python
df_movie.dtypes
```

```python
df_movie.loc[:, 'duration'] = df_movie['duration'].apply(lambda x: int(x.split(' ')[0]) if pd.notnull(x) else None)
```

```python
df_movie
```

```python
df_movie[df_movie['duration'] == '0']
```

```python
df_movie.loc[:, 'minute'] = df_movie['duration'].str.split(expand=True)[0].astype(int)
```

```python
df_movie
```

```python
df_movie['date_added'].str.split(',', expand=True)[1]
```

```python
df_movie.loc[:, 'year'] = df_movie['date_added'].str.extract('(\d{4})')
```

```python
df_movie
```
